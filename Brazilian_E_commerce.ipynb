```python
avg_pass_limit_time= (df_order_pass_Limit_date["order_delivered_carrier_date"] - df_order_pass_Limit_date["shipping_limit_date"]).mean()
avg_pass_limit_time
```

```python
order_delayed_passed_limit = order_delayed[order_delayed_and_pass_limit_date]
order_delayed_passed_limit["new_arrive_date"] = order_delayed_passed_limit["order_estimated_delivery_date"] + avg_pass_limit_time
```
